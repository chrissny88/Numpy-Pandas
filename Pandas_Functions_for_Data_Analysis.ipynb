16. Filling the Null values by central tendency (Mean, Mode, Median)

We can fill the null values with different approaches depending on the type of data we have. For categorical data, it is better to use Mode while for numerical values we will be using mean and median. Here we have filled the null values with mode for the categorical columns ‘club’ and ‘joined’.

categorical = [“Joined”, “Club”]

df[categorical]=df[categorical].fillna(df.mode().iloc[0])

```python
categorical=['joined','club_name']
fifa[categorical]=fifa[categorical].fillna(fifa.mode().iloc[0])
```
